```python
# syft absolute
import syft as sy
```

```python
node = sy.orchestra.launch(
    name="test",
    dev_mode=True,
    reset=True,
    local_db=True,
    n_consumers=4,
    create_producer=True,
)
```

```python
domain_client = node.login(email="info@openmined.org", password="changethis")
# domain_client = sy.login(email="info@openmined.org", password="changethis", port=8080)
```

```python
@sy.syft_function()
def subjob(domain):
    # stdlib
    import time

    n_iters = 10
    domain.init_progress(n_iters=n_iters)
    for i in range(n_iters):
        time.sleep(1)
        print(f"Subjob Iter {i}")
        domain.increment_progress()
    return "Done"
```

```python
domain_client.code.submit(subjob)
```

```python
@sy.syft_function_single_use()
def job(domain):
    # stdlib
    import time

    n_iters = 2
    domain.init_progress(n_iters=n_iters)
    for i in range(n_iters):
        _ = domain.launch_job(subjob)
        time.sleep(0.1)
        print(f"Job Iter {i}")
        domain.increment_progress()
    return "Done"
```

```python
domain_client.code.request_code_execution(job)
```
