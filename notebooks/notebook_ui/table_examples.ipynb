```python
# third party
import numpy as np

# syft absolute
from syft.service.action.action_object import ActionObject
from syft.service.user.user import User
```

```python
ao_1 = ActionObject.from_obj(10.0)
ao_2 = ActionObject.from_obj(20.0)
np_ao = ActionObject.from_obj(np.array([10, 20]))
user_1 = User(email="x@y.z")
user_2 = User(email="a@b.c")

# Makes table
homogenous_ao = ([ao_1, ao_2], str)
non_homogenous_same_repr = ([ao_1, ao_2, np_ao], str)
homogenous_user = ([user_1, user_2], str)
empty = ([], str)

# Doesn't make table
non_homogenous_different_repr = ([ao_1, ao_2, user_1, user_2], type(None))
non_syft_obj_1 = ([1, ao_1, ao_2], type(None))
non_syft_obj_2 = ([ao_1, ao_2, 1], type(None))

test_cases = [
    empty,
    homogenous_ao,
    non_homogenous_same_repr,
    homogenous_user,
    non_homogenous_different_repr,
    non_syft_obj_1,
    non_syft_obj_2,
]

for test_case in test_cases:
    obj, expected = test_case

    assert isinstance(obj._repr_html_(), expected)
    assert isinstance(dict(enumerate(obj))._repr_html_(), expected)
    assert isinstance(set(obj)._repr_html_(), expected)
    assert isinstance(tuple(obj)._repr_html_(), expected)
```

```python
for test_case in test_cases:
    obj, expected = test_case

    display(obj)
    print()
```

```python
for test_case in test_cases:
    obj, expected = test_case

    display(dict(enumerate(obj)))
    print()
```

```python
for test_case in test_cases:
    obj, expected = test_case

    display(set(obj))
    print()
```

```python
for test_case in test_cases:
    obj, expected = test_case

    display(tuple(obj))
    print()
```
